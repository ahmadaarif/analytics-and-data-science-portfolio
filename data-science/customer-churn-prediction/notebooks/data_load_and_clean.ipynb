```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("telco_customer_churn.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=["TotalCharges"]).copy()
```

```python
df["Churn"] = df["Churn"].map({"Yes": 1, "No": 0})
df["Churn"].value_counts(normalize=True)
```

```python
target = "Churn"
customer_id = "customerID"

numeric_features = df.select_dtypes(include=["int64", "float64"]).columns.drop(target)
categorical_features = df.select_dtypes(include=["object"]).columns.drop(customer_id)

numeric_features, categorical_features
```

- Loaded and inspected the Telco Customer Churn dataset
- Fixed data type issues and handled missing values
- Defined churn as a binary target variable
- Identified numerical and categorical features
- Created a clean dataset for EDA and modelling
